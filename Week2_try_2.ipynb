```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv('C:\\Users\\pani_puri\\Downloads\\Summer Analytics Assignment\\Week 2 hackathon\\hacktrain.csv')
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
train['class_encoded'] = encoder.fit_transform(y = train['class'])
#label_dict = dict(zip(encoder.classes_, list(encoder.transform(encoder.classes_))))
#print(label_dict)
train.head()
```

```python
id = train['ID']
train.drop(columns = ['Unnamed: 0','ID','class'],inplace = True)
train.head()
```

Feature Creation

```python
# Interpolate missing values in a cubic manner since the data has missing values and 
# contains a lot of noise
# Use pandas' interpolate method for DataFrames
#There is an interpolate function in scipy too with a different name but that is for 
#1-D arrays. Prefer using pandas function for dataframes
#'forward': Only fill missing values forward (using previous values).
# 'backward': Only fill missing values backward (using next values).
# 'both': Fill missing values in both directions.

train_interpolated = train.interpolate(method='linear', axis=0, limit_direction='both')
train_interpolated.head()
```

```python
test_result  = train['class_encoded']
train_interpolated.drop(columns = ['class_encoded'],inplace= True)
```

```python
#To reduce noise you can smoothen up data by using mean or median in a rolling window
#For spectral data there is an advanced filter called Savitzkay-Golay Filter.

from scipy.signal import savgol_filter

# Apply Savitzky-Golay filter along the rows (axis=1)
# window_length must be odd and less than or equal to the number of columns
# polyorder should be less than window_length
window_length = 5  # Adjust as needed, must be odd and <= number of columns
polyorder = 2      # Adjust as needed, must be < window_length

train_smt = pd.DataFrame(
    savgol_filter(train_interpolated.values, window_length=window_length, polyorder=polyorder, axis=1),
    columns=train_interpolated.columns,
    index=train_interpolated.index
)

train_smt.head()
```

```python
for col in train_smt.columns:
    new_col = str(col).replace('_N', '')
    train_smt = train_smt.rename(columns={col: new_col})
    
train_smt.columns = pd.to_datetime(train_smt.columns)
train_smt.head()
```

```python
#min and max of a row
features = pd.DataFrame({
    'row_min': train_smt.min(axis=1),
    'row_max': train_smt.max(axis=1)
})
features['min/max'] = features['row_min']/features['row_max']
#mean of the column
features['mean'] = train_smt.mean(axis= 1)
features['Standard_deviation'] = train_smt.std(axis = 1)

NDVI_THRESHOLD = 3500
#print(train_smt.min().min(), train_smt.max().max()) The values are scaled up so can't be sure about threshold

#So like range ke mid mein ja skte ho ek threshold ke liye
# Assuming NDVI columns are all columns in train_smt (adjust if needed)
features['days_ndvi_above_3500'] = (train_smt > NDVI_THRESHOLD).sum(axis=1)
features['days_ndvi_below_3500'] = (train_smt < NDVI_THRESHOLD).sum(axis=1)

max_slope = train_smt.diff(axis=1).max(axis=1)
min_slope = train_smt.diff(axis=1).min(axis=1)
features['max_slope_increase'] = max_slope
features['min_slope_decrease'] = min_slope

# The shape of the distribution can also provide 
# various insight about the distribution of the NVDI values
from scipy.stats import skew, kurtosis

features['skewness'] = train_smt.apply(skew, axis=1)
features['kurtosis'] = train_smt.apply(kurtosis, axis=1)

features.drop(columns=['row_max','row_min'],inplace = True)

features.head()
```

```python
# ...existing code...

# Integral/Area Under Curve: Sum of all positive NDVI values (or np.trapz)
features['ndvi_sum_positive'] = (train_smt * (train_smt > 0)).sum(axis=1)
features['ndvi_trapz'] = train_smt.apply(lambda row: np.trapz(row, dx=1), axis=1)

# Number of peaks/troughs: Using scipy.signal.find_peaks
from scipy.signal import find_peaks

def count_peaks(row):
    peaks, _ = find_peaks(row)
    return len(peaks)

def count_troughs(row):
    troughs, _ = find_peaks(-row)
    return len(troughs)

features['num_peaks'] = train_smt.apply(count_peaks, axis=1)
features['num_troughs'] = train_smt.apply(count_troughs, axis=1)

# Range/Amplitude: Max_NVDI - Min_NVDI
features['ndvi_range'] = train_smt.max(axis=1) - train_smt.min(axis=1)

# ...existing code...
```

```python
#Winter - 12 1 2 
#Spring - 3 4 5 
#Summer - 6 7 8
#Autumn - 9 10 11

# Select columns corresponding to summer months (June, July, August)
# Ensure columns are datetime

col = train_smt.columns
# Select columns where the month is June, July, or August
summer_columns = [c for c in col if c.month in [6, 7, 8]]
Winter_columns = [c for c in col if c.month in [11, 1,2]]
Spring_columns = [c for c in col if c.month in [3, 4, 5]]
Autumn_columns = [c for c in col if c.month in [9, 10, 11]]

features['Summer'] = train_smt[summer_columns].mean(axis=1)**2
features['Spring'] = train_smt[Spring_columns].mean(axis=1)**2
features['Autumn'] = train_smt[Autumn_columns].mean(axis=1)**2
features['Winter'] = train_smt[Winter_columns].mean(axis=1)**2

#Growth Seasons
features['Spring-Winter'] = features['Spring'] - features['Winter']
features['Autumn-Summer'] = features['Autumn'] - features['Summer']
features['Summer-Spring'] = features['Summer'] - features['Spring']

#Ratio of peak to dormant seasons
#features['Summer/Winter'] = features['Summer']/features['Winter']

features.head()
```

```python
from sklearn.preprocessing import StandardScaler
Scalor = StandardScaler()
features_scaled = Scalor.fit_transform(features)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(features_scaled,test_result)
```
